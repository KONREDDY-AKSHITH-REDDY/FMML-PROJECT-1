```python
#Question 1 :

def reverse_and_add_strings(A, B):
    reversestring_A = A[::-1]
    result =  reversestring_A + B

    return result # Returning the result

string_A = " HEY "
string_B = "ABHI"
Final_String = reverse_and_add_strings(string_A, string_B)
print(Final_String)
```

```python
#Question 2 :

def calculate_pay(work_hours):
    max_hours = 0
    female_worker = None

    for name, hours, gender in work_hours:
        if gender == 'F' and hours > max_hours:
            max_hours = hours
            female_worker = name

    if female_worker is not None:
        payment = max_hours * 20
        print(f"The female worker worked the most hours is {female_worker}")
        print(f"The amount she has to get paid is ${payment}")
    else:
        print("No female workers found in the list.")

workhours = [('rajesh', 100, 'F'), ('ramesh', 400, 'M'), ('Cassie', 800, 'F'), ('Maggi', 600, 'F'),('Alex', 500, 'M'), ('Raj', 225, 'M'), ('Penny', 920, 'F'), ('Ben', 300, 'M')]

calculate_pay(workhours)
```

```python
#Question 3 :
#a)Create a 5x5 matrix of the following form,


import numpy as np
base_matrix = np.array([[1, 1], [2, 2]])
horizontal_stack = np.hstack([base_matrix] * 3)
vertical_stack = np.vstack([horizontal_stack] * 3)
result_matrix = vertical_stack[:5, :5]
print(result_matrix)
```

```python
#Question 3 :
#b) find dot product of the matrix with any matrix. (Figure out the size/ shape of the matrix)

import numpy as np
A = np.array([[2, 6], [5, 1]])
B = np.array([[3, 2], [4, 9]])

if A.shape[1] == B.shape[0]:
    C = np.dot(A, B)
    print("Matrix C (Dot Product of A and B):\n", C)
else:
    print("Matrix dimensions are not compatible for dot product.")
```
