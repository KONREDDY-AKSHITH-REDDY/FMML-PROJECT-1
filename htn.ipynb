```python
#PERFECT SQUARES
def binary(s,f,a,k):
    mid=(s+f)//2
    if a[mid]*a[mid]<k and a[mid+1]*a[mid+1]>k:
        return a[mid]
    elif s==f or s>f:
        return False
    else:
        return binary(mid+1,f,l,k) or binary(s,mid-1,l,k)

k=int(input("Enter the key to search:"))
l=[1*i for i in range(k+1)]
print(binary(0,len(l)-1,l,k))
```
